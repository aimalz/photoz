## Photo-z point estimation for SDSS galaxies with LASSO, KNN, Random Forest

```python
import pyfits as pf
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
%pylab inline
import copy
```

```python
DATA = pf.open("data.fits")[1].data
```

```python
Y = DATA['redshift']
ug= DATA['dered_u'] - DATA['dered_g']
gr= DATA['dered_g'] - DATA['dered_r']
ri= DATA['dered_r'] - DATA['dered_i']
iz= DATA['dered_i'] - DATA['dered_z']
i = DATA['dered_i']
ga = DATA['dered_g'] + 5*np.log(DATA['petroRad_g'])
ra = DATA['dered_r'] + 5*np.log(DATA['petroRad_r'])
ia = DATA['dered_i'] + 5*np.log(DATA['petroRad_i'])

X = np.vstack([ug,gr,ri,iz,i,ga,ra,ia]).T
##print min(DATA["dered_u"]) , max(DATA["dered_z"]), max(gr), max(DATA["dered_u"])
X = X[(DATA["redshift"] < .6)&(DATA["dered_u"] > -9999)&(DATA["dered_g"] > -9999)&(DATA["dered_r"] > -9999)&(DATA["dered_i"] > -9999)&(DATA["dered_z"] > -9999)]
Y = Y[(DATA["redshift"] < .6)&(DATA["dered_u"] > -9999)&(DATA["dered_g"] > -9999)&(DATA["dered_r"] > -9999)&(DATA["dered_i"] > -9999)&(DATA["dered_z"] > -9999)]
X = copy.copy(X)
print np.array(X).shape , Y.shape
```

```python
pd.set_option('display.max_columns', None)
%pylab inline
import seaborn as sns
sns.set()
```

```python
import matplotlib as mpl
import matplotlib.cm as cm
norm = mpl.colors.Normalize(vmin=min(Y), vmax=max(Y))
cmap = cm.jet_r
L = pd.DataFrame(X[0:2000] , columns=["$u-g$","$g-r$","$r-i$","$i-z$","$i$","$g+5log(\ttheta_{g})$","$r+5log(\ttheta_{r})$","$i+5log(\ttheta_{i})$"])
m = cm.ScalarMappable(norm=norm, cmap=cmap)
rez = pd.scatter_matrix(L,alpha=0.5,figsize=[15,15],color=m.to_rgba(Y))
```

Now, we split the data into training and test sets:

```python
half = int(len(Y)/2)
from sklearn import preprocessing
X_scaled = preprocessing.scale(X) # many methods work better on scaled X
train_X = X_scaled[:half,:]
train_Y = Y[:half]
test_X = X_scaled[half:,:]
test_Y = Y[half:]
```

#### Regression with Lasso Objective function Least-square + L1 penalty(sparisty prior)

```python
from sklearn import linear_model
LR = linear_model.Lasso(alpha = 10**-6. , max_iter=10000000, tol=0.000001)
```

fitting

```python
LR.fit(train_X,train_Y)
```

making prediction for the validation set

```python
Y_lr_pred = LR.predict(test_X)
from sklearn.metrics import mean_squared_error
mse = np.sqrt(mean_squared_error(test_Y,Y_lr_pred)) ; print "MSE",mse
```

```python
plot(test_Y,Y_lr_pred,'o',alpha=0.1)
title("Lasso Regression photoz prediction - MSE = %.8f" % mse)
xlabel("Spectroscopic Redshift")
ylabel("Predicted photo-z")
xlim(min(test_Y),.6)
ylim(min(test_Y),.6)
plot(test_Y , test_Y ,color="red")
figsize=(10,10)
```

##### k nearest neighbor regression

```python
from sklearn import neighbors
clf = neighbors.KNeighborsRegressor(10)
train_X = X_scaled[:half]
test_X = X_scaled[half:]
clf.fit(train_X,train_Y)
```

```python
Y_knn_pred = clf.predict(test_X)
mse = np.sqrt(mean_squared_error(test_Y,Y_knn_pred)) ; print "MSE",mse
```

```python
plot(test_Y,Y_knn_pred,'o',alpha=0.1)
title("kNN photoz prediction - MSE = %.8f" % mse)
xlabel("Spectroscopic Redshift")
ylabel("Predicted photo-z")
xlim(min(test_Y),.6)
ylim(min(test_Y),.6)
plot(test_Y , test_Y ,color="red")
figsize=(10,10)
```

#### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=10, n_jobs=4)
rf.fit(train_X,train_Y)
```

```python
Y_rf_pred = rf.predict(test_X)
mse = np.sqrt(mean_squared_error(test_Y,Y_rf_pred)) ; print "MSE",mse
```

```python
plot(test_Y,Y_rf_pred,'o',alpha=0.1)
title("RF photoz prediction - MSE = %.8f" % mse)
xlabel("Spectroscopic Redshift")
ylabel("Predicted photo-z")
xlim(min(test_Y),.6)
ylim(min(test_Y),.6)
plot(test_Y , test_Y ,color="red")
figsize=(10,10)
```
